```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# xのデータ
x = np.array([i for i in range(100)])
```

```python
# yのデータ 2 * i + 10 + i * sin(i)     wx + b
y = np.array([2 * i + 10 + i * math.sin(i) for i in range(100)])
```

```python
# 形状 次元
print(x.shape, x.ndim)
print(y.shape, x.ndim)
```

```python
# 行数:-1=自由  列数:1
x = x.reshape((-1, 1))
y = y.reshape((-1, 1))
```

```python
# 線形回帰
# 二乗和誤差の最小化
reg = LinearRegression().fit(x, y)
```

```python
# 係数(決定係数)  切片(回帰係数)
reg.coef_[0][0], reg.intercept_[0]
```

```python
# 元のデータ x, yの表現
plt.plot(x, y)
predict_y = [reg.coef_[0][0] * i + reg.intercept_[0] for i in range(x.shape[0])]
plt.plot(x, predict_y)
```

```python
plt.plot(x, reg.predict(x))
```

```python
# 精度はあまり良くない（当然）
reg.score(x, y)
```
